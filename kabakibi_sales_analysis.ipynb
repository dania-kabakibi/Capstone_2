```python
prod = pd.read_csv('Products.csv')

print("=== Products info ===")
prod.info()
print("\n=== First three rows ===")
prod.head(3)
```

```python
store_det = pd.read_csv('StoreDetail.csv')

print("=== StoreDetail info ===")
store_det.info()
print("\n=== First three rows ===")
store_det.head(3)
```

```python
store_sales = pd.read_csv('StoreSales.csv')

print("=== StoreSales info ===")
store_sales.info()
print("\n=== First three rows ===")
store_sales.head(3)
```

## Core Marketing Analysis

### 1. Assigned Territories Metadata

#### - The territory managers for the sales territories assigned (Maine, Maryland).

```python
maine_manager = store_det.loc[store_det['State'] == 'Maine', 'Territory Manager'].unique()[0]
maryland_manager = store_det.loc[store_det['State'] == 'Maryland', 'Territory Manager'].unique()[0]

print(f"The Territory Manager for Maine is: {maine_manager}")
print(f"The Territory Manager for Maryland is: {maryland_manager}")
```

#### - The store IDs and cities for the stores in each assigned sales territory.

```python
maine_stores = store_det.loc[store_det['State'] == 'Maine', ['Store ID', 'Store Location']]
maryland_stores = store_det.loc[store_det['State'] == 'Maryland', ['Store ID', 'Store Location']]

print("Maine Stores:")
print(maine_stores)
print("\nMaryland Stores:")
print(maryland_stores)
```

---
### 2. Monthly Revenue Trends (Maine vs. Maryland)

#### Monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data.

```python
# Merge and convert date to datetime
mer_sd_ss = store_det.merge(store_sales, on='Store ID')
mer_sd_ss['Transaction Date'] = pd.to_datetime(mer_sd_ss['Transaction Date'])

# Create the monthly period Series
m = mer_sd_ss['Transaction Date'].dt.to_period('M')

# Isolate the data for each territory
maine = mer_sd_ss.loc[mer_sd_ss['State'] == 'Maine']
maryland = mer_sd_ss.loc[mer_sd_ss['State'] == 'Maryland']

# Group by the monthly period to get territory totals
maine_monthly_revenue = maine.groupby(m)['Sale Amount'].sum()
maryland_monthly_revenue = maryland.groupby(m)['Sale Amount'].sum()

print("--- Maine Monthly Total Revenue ---")
print(maine_monthly_revenue.to_frame('Total Revenue'))

print("\n--- Maryland Monthly Total Revenue ---")
print(maryland_monthly_revenue.to_frame('Total Revenue'))
```

---
### 3. Store Performance Leaderboards

#### The sales performance ranking of each store in each sales territory.

```python
# Calculate total revenue per store and sort from highest to lowest
maine_ranking = maine.groupby(['Store ID', 'Store Location'])['Sale Amount'].sum().sort_values(ascending=False)
maryland_ranking = maryland.groupby(['Store ID', 'Store Location'])['Sale Amount'].sum().sort_values(ascending=False)

print("--- Maine Store Sales Ranking ---")
print(maine_ranking.to_frame('Total Revenue'))

print("\n--- Maryland Store Sales Ranking ---")
print(maryland_ranking.to_frame('Total Revenue'))
```

---
### 4. Top-Spending Rewards Customers

```python
# Drop rows where RewardsID is missing and convert to int
sales_with_rewards = mer_sd_ss.dropna(subset=['RewardsID']).copy()
sales_with_rewards['RewardsID'] = sales_with_rewards['RewardsID'].astype(int)

# Merge customer list with the sales data
mer_cl_ss = pd.merge(cust_list, sales_with_rewards, left_on='cust_id', right_on='RewardsID')

# Isolate the data for each territory
maine_cust = mer_cl_ss.loc[mer_cl_ss['State'] == 'Maine']
maryland_cust = mer_cl_ss.loc[mer_cl_ss['State'] == 'Maryland']

# Group by customer ID and name, sum their spending, and sort from highest to lowest
maine_top_customers = maine_cust.groupby(['cust_id', 'name'])['Sale Amount'].sum().sort_values(ascending=False)
maryland_top_customers = maryland_cust.groupby(['cust_id', 'name'])['Sale Amount'].sum().sort_values(ascending=False)

# Display the top 5 customers for each territory
print("--- Top Customers in Maine ---")
print(maine_top_customers.head().to_frame('Total Spend'))

print("\n--- Top Customers in Maryland ---")
print(maryland_top_customers.head().to_frame('Total Spend'))
```

---
### 5. Product Category (Number of transactions vs. Revenue)

#### - The number of transactions per month by product category in each assigned territory.

```python
# Merge products with store sales
mer_p_ss = prod.merge(store_sales, on='Prod Num')
# Merge on both shared columns to keep the DataFrame clean
mer_pc_p_ss = prod_cate.merge(mer_p_ss, on=['CategoryID', 'SubcategoryID'])

# Merge StoreDetail with the (ProductCategories, Products, and StoreSales)
full_df = store_det.merge(mer_pc_p_ss, on='Store ID')

# Convert transaction dates to monthly periods
full_df['Transaction Date'] = pd.to_datetime(full_df['Transaction Date'])
mon = full_df['Transaction Date'].dt.to_period('M')

# Isolate the data for each assigned territory
maine_data = full_df.loc[full_df['State'] == 'Maine']
maryland_data = full_df.loc[full_df['State'] == 'Maryland']

# Group by Category and Month, then count the transactions
maine_num_of_trans = maine_data.groupby(['Category', mon])['Sale Amount'].count()
maryland_num_of_trans = maryland_data.groupby(['Category', mon])['Sale Amount'].count()

print("--- Maine: Monthly Transactions by Category ---")
print(maine_num_of_trans.to_frame('Number of Transactions'))

print("\n--- Maryland: Monthly Transactions by Category ---")
print(maryland_num_of_trans.to_frame('Number of Transactions'))
```

#### - Total sales revenue per month by category

```python
# Group by Month and Category for Maine
maine_monthly_revenue = maine_data.groupby([mon, 'Category'])['Sale Amount'].sum()

# Group by Month and Category for Maryland
maryland_monthly_revenue = maryland_data.groupby([mon, 'Category'])['Sale Amount'].sum()

print("--- Maine: Monthly Revenue by Category ---")
print(maine_monthly_revenue.to_frame('Total Revenue'))

print("\n--- Maryland: Monthly Revenue by Category ---")
print(maryland_monthly_revenue.to_frame('Total Revenue'))
```
